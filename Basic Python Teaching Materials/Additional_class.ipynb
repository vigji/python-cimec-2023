```python
# A0, A1, ... B0, B1, ... C9

groups = ["A", "B", "C"]
n_mice = 10


mice = []

for g in groups:
  for i in range(n_mice):
    id = g + "_" + str(i)
    mice.append(id)
```

```python
mice
```

```python
alive_mice = []

for i in range(len(mice)):
  if i % 3 != 0:
    alive_mice.append(mice[i])
    
alive_mice
```

```python
alive_mice = []
i = 0
for m in mice:
  if i % 3 != 0:
    alive_mice.append(m)
  i += 1

alive_mice
```

```python
alive_mice = []

for i, mouse in enumerate(mice):
  if i % 3 != 0:
    alive_mice.append(mouse)

alive_mice
```

```python
def generate_ids(groups, n_mice):
  mice = []
  for g in groups:
    for i in range(n_mice):
      id = g + "_" + str(i)
      mice.append(id)
  return mice

groups = ["A", "B"]

generate_ids(groups, 5)
```

```python
def kill_some_mice(mice, n):
  mice_alive = []
  for i, mouse in enumerate(mice):
    if i % n != 0:
      mice_alive.append(mouse)
  return mice_alive
```

```python
groups = ["A", "B", "C"]
n_mice = 7

data = generate_ids(groups, n_mice)
data = kill_some_mice(data, 3)
data
```

```python
generate_ids(groups, n_mice)
```
