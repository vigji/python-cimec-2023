```python
#Exercise 11:
# Write a function that will take input string representing genetic sequence and
# will give a complementary string as an output

dna_sequence = "ATGCTATGCGCGCGTAGCGTG"
def complementary_sequence(sequence):
  compl_sequence = ""
  for n in sequence:
    if n == 'A':
       compl_sequence += 'T'
    elif n == 'T':
      compl_sequence += 'A'
    elif n == 'G':
      compl_sequence += 'C'
    elif n == 'C':
      compl_sequence += 'G'
  return compl_sequence

complementary = complementary_sequence(dna_sequence)
print(complementary)

# alternative solution
def complementary_seq_2(sequence):
  conversion_dict = {"A": "T", "G": "C", "T": "A", "C": "G"}
  other_strand = ""
  for base in sequence:
    other_strand += conversion_dict[base]
  return other_strand

print(complementary_seq_2(dna_sequence))

# Another alternative solution

def complementary_seq_3(sequence):
    primary_sequence = "ATGC"
    secondary_sequence = "TACG"
    complementary_strand = ""
    for base in sequence:
      complementary_strand += secondary_sequence[primary_sequence.find(base)]
    return complementary_strand

print(complementary_seq_3(dna_sequence))
```

```python
# Exercise 4

# create a function that takes two iterables and creates a list of all possible 
# pairings of their contents, e.g. 
# "abc" and "def" will give "ab", "bd", "cd", "ae", "af", "be", "bf", "cd", "ce", "cf"
# There's a built-in which does this, look at itertools

def combinations(iter1, iter2):
    output_list = []
    for i in iter1:
        for j in iter2:
            output_list.append(i+j)
    return output_list

print(combinations("abc", "def"))
```

```python
# Exercise 10:
# Let's imagine (again) that "spikes" contain some processed neural signal recorded over 
# some time, whenever signal reaches 1, we have a spike. 
# Create a function that will take as an input spiking data and a size of a window, and 
# calculate spiking frequency (number of spikes/frame) in 
# each window of size specified by a variable. Save the output of the function to a variable 
# called firing_rate
# E.g. you divide the list of 1000 data points in frames of 10, and for every 10 frames 
# you calculate number of spikes; as an output you have list 
# of 1000/10 = 100 elements, where every number represents firing rate in the window of 10 frames

import numpy as np
import matplotlib.pyplot as plt
import statistics
np.random.seed(9)

spikes = np.random.choice(np.array([0,1]), size= 1000, p=[0.9, 0.1])  # here we generate the spikes signal
plt.plot(spikes);

# Your code goes here
def spiking_freq(spike_data, window_size):
  frequency = []
  for i in range(0,len(spike_data), window_size):
    frequency.append(sum(spike_data[i:i+window_size])/window_size)
  return frequency

firing_rate = spiking_freq(spikes, 10)


# Plotting(uncomment to run)
plt.figure()
plt.bar(range(0, len(firing_rate)), firing_rate);
```

```python
#Exercise 12:
# Modify a function from the previous exercise so that it will provide a homologous RNA or 
# DNA string depending on additional boolean variable called "dna" 
# (if dna == False, we need to produce RNA sequencee)

dna_sequence = "ATGCTATGCGCGCGTAGCGTG"

def complementary_sequence(sequence, dna):
  compl_sequence = ""
  for n in sequence:
    if (n == 'A') and (dna == True):
       compl_sequence += 'T'
    elif (n == 'A') and (dna == False):
       compl_sequence += 'U'
    elif n == 'T':
      compl_sequence += 'A'
    elif n == 'G':
      compl_sequence += 'C'
    elif n == 'C':
      compl_sequence += 'G'
  return compl_sequence

complementary = complementary_sequence(dna_sequence, False)
print(complementary)

# alternative solution
def complementary_seq_2(sequence, is_dna):
    conversion_dict = {"A": "T", "G": "C", "T": "A", "C": "G"}
    if not is_dna:
        conversion_dict.update({"A": "U"})
    other_strand = ""
    for base in sequence:
      other_strand += conversion_dict[base]
    return other_strand

# Another alternative solution

def complementary_seq_3(sequence, is_dna):
    primary_sequence = "ATGC"
    if is_dna:
        secondary_sequence = "TACG"
    else:
        secondary_sequence = "UACG"
    complementary_strand = ""
    for base in sequence:
      complementary_strand += secondary_sequence[primary_sequence.find(base)]
    return complementary_strand

print(complementary_seq_2(dna_sequence, False))
print(complementary_seq_3(dna_sequence, False))
```
