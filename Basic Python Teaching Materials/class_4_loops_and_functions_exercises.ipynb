```python
# Exercise 11: Write a function that calculates a running average of a signal within a given window of 10  using for loop and if (or solving the problem of the first 5 and last 5 samples in the other way)
import numpy as np
import matplotlib.pyplot as plt

random_seed = 9
random_number_generator = np.random.default_rng(random_seed)

mu, sigma = 0, 0.01 
signal = random_number_generator.normal(mu, sigma, 200) # here we generate the signal by drawing samples from normal distribution
plt.plot(signal); # plot of the signal
# print(signal)
# Your function goes here
def running_average(input_signal, window):
  temp_list = []
  half_window = int(window / 2)
  smoothed_x = range(half_window, len(input_signal)-half_window)
  for i in smoothed_x:
    current_window_mean = sum(input_signal[i-half_window: i+half_window]) / window
    temp_list.append(current_window_mean)
  return smoothed_x, temp_list


# Use your running average and assign it to a variable named signal_smoothed
x_values, signal_smoothed = running_average(signal, 50)

# Plot filtered signal (uncomment to plot)
plt.plot(x_values, signal_smoothed);
```
