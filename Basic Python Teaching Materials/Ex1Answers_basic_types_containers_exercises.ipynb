```python
# Exercise 1 - basic maths

# Calculate the mean of these numbers, and assign to a variable: 346, 23, 116, 89, 150.4, 12

# Solution 1:

x = 346 + 23 + 116 + 89 + 150.4 + 12
print (x) # You can always use print to check everything is going as you expect

mean_num = x / 6
print('The mean of these numbers are', mean_num)

# Solution 2:

num = [346, 23, 116, 89, 150.4, 12]

total = sum(num)
print(total)

mean_num2 = total / len(num)
print (f'The mean of these numbers are {mean_num2}')

# Solution 3:

from statistics import mean

mean_num3 = mean(num)
print('The mean of these numbers are {}.'.format(mean_num3))


# Calculate the mean cubed and add 1280000, then calculate the square root, and assign to a variable

# Solution 1
cubed = mean_num * mean_num *mean_num
print('The cube of {} is {}'.format(mean_num, cubed))

add_num = cubed + 1280000
print('When you add 1280000 to {:.2f}, you get {:.2f}.'.format(cubed, add_num)) # I get two decimals from cubed
print('The add_num is', add_num)

dec = round(add_num, 2)

square_root = dec ** 0.5
print(f'The square root of {dec} is {square_root}.')

# Solution 2

import math

cubed_num = mean_num ** 3
addition = cubed_num + 1280000
sqr_root = math.sqrt(addition)
decimal = round(sqr_root, 2)
print('The answer is', decimal)

# Determine the integer and remainder from the following calculation, and assign to variables: 12 divided by 5

# Solution 1

integer = 12 // 5
remainder = 12 % 5

# Solution 2

integer = int(12 / 5)
remainder = 12 - (integer * 5)

print(integer)
print(remainder)
```

```python
# Exercise 4 - Methods
# Remember! A method is just a function attached to an object

test_string = "This is a test string, it has almost everything you'd expect: CAPITAL LETTERS, p.u,n/c%tuation, some numbers... 1, 67, 128, 1000; as well as somewhere to insert something else: {}"
# Using dir, identify 3 string methods and apply them to the string above (ignore methods beginning with __ for now).
# As an extra exercise, try to work out what you expect first, and then assign that to expected output
# Then test your changed output with your expected (using ==) and see if you predicted correctly!

# Methods we've talked about previously:
# upper, lower, format, find, count, endswith, startswith
# Other suggestions
# split, rstrip (or lstrip)

expected_output1 = "capital"
a_word = test_string.find('CAPITAL')
an_output = test_string.casefold()[a_word : (a_word+7)]

print(an_output == expected_output1)



expected_output2 = 1000
word_two = test_string.find(' 1000;')
output_2 = test_string[word_two : (word_two+6)]
output_two = output_2.lstrip().rstrip(";")

print(int(output_two) == expected_output2)



expected_output3 = ''

new_sentence = test_string.format(",")
output_three = new_sentence.split(',')

print(output_three[-1] == expected_output3)
```
